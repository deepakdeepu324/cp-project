```python
#Employee Attrition Analysis
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Load Dataset 
# Note: You can use the IBM HR Analytics Attrition Dataset (Kaggle) for this.
# For this example, we create a small synthetic structure for demonstration.
data = {
    'Age': [34, 45, 23, 50, 31, 22, 40, 29],
    'OverTime': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes'],
    'JobSatisfaction': [1, 4, 2, 4, 3, 1, 4, 2], # 1-Low, 4-High
    'YearsAtCompany': [2, 10, 1, 20, 5, 1, 12, 3],
    'Attrition': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)

# 2. Data Preprocessing & Categorical Encoding
# Algorithms need numbers, so we convert "Yes/No" to 1/0
le = LabelEncoder()
df['OverTime'] = le.fit_transform(df['OverTime'])
df['Attrition'] = le.fit_transform(df['Attrition'])

# 3. Splitting Features and Target
X = df.drop('Attrition', axis=1)
y = df['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 5. Implementing Classification (Decision Tree)
# We use 'entropy' to measure the quality of the split
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X_train, y_train)

# 6. Prediction and Evaluation
y_pred = model.predict(X_test)

print(f"Model Accuracy: {accuracy_score(y_test, y_pred) * 100}%")
print("\n--- Detailed Classification Report ---")
print(classification_report(y_test, y_pred))

# 7. Interpreting Feature Importance
# This shows which factors (like OverTime) drive attrition most
importance = pd.Series(model.feature_importances_, index=['Age', 'OverTime', 'JobSatisfaction', 'YearsAtCompany'])
print("\n--- Key Attrition Drivers ---")
print(importance.sort_values(ascending=False))
```
